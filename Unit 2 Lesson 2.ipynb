```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('14zpallagi.csv')
#agi = size of adjusted gross income
#1 = $1 to $24,999
#2 = $25,000 to $49,999
#3 = $50,000 to $74,999
#4 = $75,000 to $99,999
#5 = $100,000 to $199,999
#6 = $200,000+
```

```python
plt.figure(figsize=(40,4))
plt.plot(df.groupby('STATE').median()['agi_stub'], color='red')
plt.show()
#This graph shows the median adjusted gross income per state.It shows that as a whole, the median AGI of most states fall
#somewhere between $50,000 to $74,999. The only exception to this is South Dakota, which has an AGI score of 4. This seems
#to indicate that the median individuals are more wealthy there than the other states.
```

```python
plt.figure(figsize=(40, 8))

plt.subplot(1, 2, 1)
plt.scatter(x = df['STATE'], y = df['mars1'], color='purple')
plt.ylabel('Individuals')
plt.title('Individual Returns')

plt.subplot(1, 2, 2)
plt.scatter(x = df['STATE'], y = df['MARS2'], color='green')
plt.ylabel('Joint')
plt.title('Joint Returns')

plt.tight_layout()
plt.show()
#These subplots show the amount of individual or joint tax returns. The data appears to show that there are more individuals
#who file jointly than individiually.
```

```python
plt.figure(figsize=(40,5))
plt.plot(df.groupby(['STATE']).sum()['N1'], color='red')
plt.show()
#This graph shows the number of total recorded tax returns (N1) from each state. This data shows that California, then Texas,
#then New York are the top three states with the most reported tax returns for the tax year 2014.
```

```python
plt.figure(figsize=(40, 8))

plt.subplot(1, 2, 1)
plt.plot(df.groupby('STATE').median()['agi_stub'], color='purple')
plt.ylabel('Adjusted gross income')
plt.title('AGI')

#A04800 is taxable income amount
plt.subplot(1, 2, 2)
plt.scatter(x = df['STATE'], y = df['A04800'], color='green')
plt.ylabel('Taxable Income')
plt.title('Taxable Income')

plt.tight_layout()
plt.show()
#The intent of the graphs here was to generate graphs showing if there is any correlation betwen the AGI and the taxable
#income. Although the median AGI for SD is highest among the states, it has one of the lowest taxable income. Texas and 
#California, two states with some of the higher taxable income, have AGI that is inline with the other states.
```
