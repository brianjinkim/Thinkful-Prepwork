```python
#Generate a random variable with 100 datapoints
#Graph onto a histogram
#Compute mean and std. deviation and plot as vertical lines
#Decide whether the statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
power = np.random.power(3, 100)
plt.hist(power)
#plotting vertical line for mean
plt.axvline(power.mean(), color='b', linestyle='solid', linewidth=2)
#plotting vertical dashed lines for std. deviation
plt.axvline(power.mean() + power.std(), color='black', linestyle='dashed', linewidth=2)
plt.axvline(power.mean() - power.std(), color='black', linestyle='dashed')
plt.show()
#The data seems to show that most of the points in the power histogram skews to the left of the mean
```

```python
logistic = np.random.logistic(5, 3, 100)
plt.hist(logistic)
#plotting vertical line for mean
plt.axvline(logistic.mean(), color='b', linestyle='solid', linewidth=3)
#plotting vertical lines for std. deviation
plt.axvline(logistic.mean() + logistic.std(), color='black', linestyle='dashed')
plt.axvline(logistic.mean() - logistic.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
#Most of the data for the logistic histogram falls between the std. deviation, and the skew is generally not as large as some of the other histograms.
```

```python
negative_binomial = np.random.negative_binomial(20, .5, 100)
plt.hist(negative_binomial)
#plotting vertical line for mean
plt.axvline(negative_binomial.mean(), color='b', linestyle='solid', linewidth=3)
#plotting vertical lines for std. deviation
plt.axvline(negative_binomial.mean() + negative_binomial.std(), color='black', linestyle='dashed')
plt.axvline(negative_binomial.mean() - negative_binomial.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
#Like the logistic histogram, most of the data falls between the std. deviations
```

```python
logseries = np.random.logseries(.5, 100)
plt.hist(logseries)
#plotting vertical line for mean
plt.axvline(logseries.mean(), color='b', linestyle='solid', linewidth=3)
#plotting vertical lines for std. deviation
plt.axvline(logseries.mean() + logseries.std(), color='black', linestyle='dashed')
plt.axvline(logseries.mean() - logseries.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
#While most of the data falls between the std. deviation for log series, there is also only one tail (right side) and appears exponential
```

```python
rayleigh = np.random.rayleigh(3, 100)
plt.hist(rayleigh)
#plotting vertical line for mean
plt.axvline(rayleigh.mean(), color='b', linestyle='solid', linewidth=3)
#plotting vertical lines for std. deviation
plt.axvline(rayleigh.mean() + rayleigh.std(), color='black', linestyle='dashed')
plt.axvline(rayleigh.mean() - rayleigh.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
#The rayleigh histogram data appears to be distorted more so on the right side. 
```

```python
vonmises = np.random.vonmises(.3, 3, 100)
plt.hist(vonmises)
#plotting vertical line for mean
plt.axvline(vonmises.mean(), color='b', linestyle='solid', linewidth=3)
#plotting vertical lines for std. deviation
plt.axvline(vonmises.mean() + vonmises.std(), color='black', linestyle='dashed')
plt.axvline(vonmises.mean() - vonmises.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
#Like nearly all the other histograms, the vonmises histogram appears to be skewed as well.
```

```python
#Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
#Add them together to create a third variable.
#Graph the third variable using a histogram.
#Compute the mean and standard deviation and plot them as vertical lines on the histogram.
#Evaluate the descriptive statistics against the data.

normal1 = np.random.normal(5, .5, 100)
normal2 = np.random.normal(10, 1, 100)
normal3 = normal1 + normal2
plt.hist(normal3)
plt.axvline(normal3.mean(), color='b', linestyle='solid')
plt.axvline(normal3.mean() - normal3.std(), color='black', linestyle='dashed')
plt.axvline(normal3.mean() + normal3.std(), color='black', linestyle='dashed')
plt.show()
#A good chunk of the data falls between the std. deviation, but does not seem to form a more uniform bell curve.
```
